```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix

# Load the dataset
df = pd.read_csv(r"C:\Users\rhkee\Downloads\healthcare-dataset-stroke-data.csv")

# Define X and y
X_logistic = df.drop('stroke', axis=1)  # Assuming 'stroke' is the target variable
y_logistic = df['stroke']

# Identify categorical and numerical columns
numerical_cols = X_logistic.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = X_logistic.select_dtypes(include=['object']).columns

# Preprocessing for numerical and categorical columns with imputation
numerical_transformer = make_pipeline(SimpleImputer(strategy='mean'), StandardScaler())
categorical_transformer = make_pipeline(SimpleImputer(strategy='most_frequent'), OneHotEncoder(handle_unknown='ignore'))

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols)
    ])

# Logistic Regression Pipeline
logistic_pipeline = make_pipeline(preprocessor, LogisticRegression(random_state=0))

# Split the data into training and testing sets
X_train_logistic, X_test_logistic, y_train_logistic, y_test_logistic = train_test_split(X_logistic, y_logistic, test_size=0.2, random_state=0)

# Train the logistic regression model
logistic_pipeline.fit(X_train_logistic, y_train_logistic)

# Predict on the test set
y_pred_logistic = logistic_pipeline.predict(X_test_logistic)

# Evaluate the model
accuracy = accuracy_score(y_test_logistic, y_pred_logistic)
conf_matrix = confusion_matrix(y_test_logistic, y_pred_logistic)
print("Logistic Regression Accuracy:", accuracy)
print("Confusion Matrix:\n", conf_matrix)
```

```python
#dummy
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_score
from matplotlib import pyplot as plt

# Load the dataset
dataset = pd.read_csv(r"C:\Users\rhkee\Downloads\healthcare-dataset-stroke-data.csv")
dataset['gender'] = dataset['gender'].map({'Male': 0, 'Female': 1})
# Display the first and last 5 rows of the dataset
print(dataset.head())
print(dataset.tail())


# Basic data visualization
# Assuming the 'stroke' column is the target variable for logistic regression
counts = []
counts.append((dataset['stroke'] == 0).sum())
counts.append((dataset['stroke'] == 1).sum())
print(counts)
stroke = ('No', 'Yes')
plt.bar(stroke, counts, color=['yellow', 'purple'])
plt.title('Bar plot of Stroke dataset')
plt.xlabel('Stroke')
plt.ylabel('Frequency')
plt.show()


# Preparing data for Logistic Regression
# Dropping non-numeric and unnecessary columns for simplicity
# You may need to handle categorical variables more carefully
X = dataset.iloc[:, 0:6].values
y = dataset['stroke'].values

# Splitting the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Training the logistic regression model on the training set
classifier = LogisticRegression(random_state=0)
classifier.fit(X_train, y_train)

# Predicting the test set results
y_pred = classifier.predict(X_test)

# Making the confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)

# Computing accuracy with k-fold cross-validation
accuracies = cross_val_score(estimator=classifier, X=X, y=y, cv=10)
print("Accuracies for each fold:")
print(accuracies)
print("Accuracy: {:.2f}%".format(accuracies.mean() * 100))
```
